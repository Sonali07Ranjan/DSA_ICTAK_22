```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Read the data into python

```python
df=pd.read_csv('week10.csv')
df.head()
```

#### 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations

```python
df[['Hardness','Sulfate','Organic_carbon']].describe()
```

```python
fig,axes=plt.subplots(nrows=1, ncols=3,figsize=(14,5))
sns.histplot(data=df['Hardness'],kde=True,stat='density',ax=axes[0])
sns.histplot(data=df['Sulfate'],kde=True,stat='density',ax=axes[1])
sns.histplot(data=df['Organic_carbon'],kde=True,stat='density',ax=axes[2])
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1, ncols=3,figsize=(8,5))
sns.boxplot(data=df['Hardness'],ax=axes[0]).set(title='Hardness boxplot')
sns.boxplot(data=df['Sulfate'],ax=axes[1]).set(title='Sulfate boxplot')
sns.boxplot(data=df['Organic_carbon'],ax=axes[2]).set(title='Organic carbon boxplot')
plt.tight_layout()
```

Observations: From the above graphs we can see that all 3 parameters Hardness,Sulfate and Organic carbon are continuous variables having an approximate normal distribution in the dataset with their means about 196, 333 and 14 respectively. From the boxplots we can observe presence of few outliers across all 3 parameters in both upper and lower regions.

#### 4. Carry out correlation analysis on the dataset

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df.corr(),annot=True,linewidth=0.2,cmap='icefire')
```

Overall correlation between the parameters can be seen as very weak

The very low negative correlation factor between ph and chloramines shows extremely weak correlation or almost zero correlation hence concludes the absence of any linear correlation between the two variables
